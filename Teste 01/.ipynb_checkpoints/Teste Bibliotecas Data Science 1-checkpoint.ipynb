<h2>CLASSIFICANDO COM KNN (K Nearest Neighbor)</h2>
<p style="text-align:justify;text-justify:inter-word;">No reconhecimento de padrões , o algoritmo k- vizinhos mais próximos (kNN) é um método supervisionado, não-paramétrico usado para classificação e regressão.  Nos dois casos, a entrada consiste nos k exemplos de treinamento mais próximos no espaço de recursos . A saída depende se kNN é usado para classificação ou regressão:</p>
<p style="text-align:justify;text-justify:inter-word;">A saída depende se k-NN é usado para classificação ou regressão: </p>
<ul style="text-align:justify;text-justify:inter-word;">
<li> Na classificação k-NN , a saída é uma associação de classe. Um objeto é classificado pelo voto de pluralidade de seus vizinhos, sendo o objeto atribuído à classe mais comum entre os k vizinhos mais próximos ( k é um número inteiro positivo , geralmente pequeno). Se k  = 1, o objeto é simplesmente atribuído à classe do único vizinho mais próximo.</li>
<li> Na regressão k-NN , a saída é o valor da propriedade para o objeto. Este valor é a média dos valores de k vizinhos mais próximos</li>
</ul>
<p>k é um inteiro positivo, tipicamente pequeno e impar.</p>
<p>Se k = 1, então o objeto é simplesmente atribuído à classe desse único vizinho mais próximo.</p>

<h2> USANDO A BIBLIOTECA SCIKIT-LEARN </h2>

```python
## importando bibliotecas
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
## Vamos amostrar rapidamente um conjunto de treinamento, 
## mantendo 40% dos dados para testar (avaliar) o classificador (validação cruzada)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
# mostrando os conjuntos de treinamento e teste
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
## usando o classificador KNN da biblioteca SciKit-Learn
knn= KNeighborsClassifier (n_neighbors=3)

## dados de treinamento
##X = df.iloc[:,0:4].values ##parte do conjunto
## classes alvo
##y = df[['target']].values.ravel()

##treinando
knn.fit(X_train,y_train)
```

<h3>TESTANDO CLASSIFICADOR</h3>

```python
## USANDO O CLASSIFICADOR TREINADO
##classificando uma nova flor com base em seus atributos
knn.predict([[6.5,5.5,4.5,1.3]])
```

```python
## USANDO O CLASSIFICADOR TREINADO
##classificando uma outra flor, também com base em seus atributos
knn.predict([[1.5,2.7,1.9,0.35]])
```

<h3> ANALISANDO E VALIDANDO OS RESULTADOS OBTIDOS</h3>

 <h4>VALIDAÇÃO CRUZADA OU CROSS VALIDATION</h4>

```python
#usando o cojunto de testes para a classificação
resultado = knn.predict(X_test)
resultado
```

<h4>MATRIZ DE CONFUSÃO</h4>

```python
## Matriz de Confusão 
pd.crosstab(y_test,resultado,rownames=['Real'],colnames=['Predito'],margins=True)
```

<p style="text-align:justify;text-justify:inter-word;">A <strong>matriz de confusão</strong> é uma tabela que mostra as frequências classificação para cada classe do modelo.</p>

<p style="text-align:justify;text-justify:inter-word;">Por exemplo, dos 23 valores que o modelo previu como sendo da classe <strong>VERSICOLOR</strong>, 22 elementos foram classificados corretamente e 1 foi classificado como sendo da classe <strong>VIRGINICA</strong>, ou seja, nesse caso o modelo errou.</p>
<p style="text-align:justify;text-justify:inter-word;">Antes de apresentar as métricas, é necessário conhecer o que é verdadeiro positivo, falso positivo, falso negativo e verdadeiro negativo.</p>

<table cellpadding='0' cellspacing='0'>
    <tr>
         <td style="color: rgb(0,0,0); background-color: rgb(153,255,153)">TP - True Positives (verdadeiros positivos)</td>
        <td style="color: rgb(0,0,0); background-color: rgb(255,153,153)">FP - False Positives (falsos positivos)</td>
    </tr>
    <tr>
        <td style="color: rgb(0,0,0); background-color: rgb(255,153,153)">FN - False Negatives (falsos negativos) </td>
        <td style="color: rgb(0,0,0); background-color: rgb(153,255,153)">TN - True Negatives (verdadeiros negativos)</td>
    </tr>    
</table> 



<br><strong>Significado de cada um desses valores:</strong>
<ul style="text-align:justify;text-justify:inter-word;">
    <li><strong>TP - </strong> Um <strong>verdadeiro positivo</strong> é um resultado em que o modelo prediz corretamente a classe positiva. Em outras palavras, indica a quantidade de registros que são classificados como positivos corretamente, ou seja, a resposta do classificador apontou que o item faz parte da classe e o item realmente faz parte da classe.</li>
    <li><strong>FP - </strong> Um <strong>falso positivo</strong> é um resultado em que o modelo prediz incorretamente a classe positiva. Neste caso,  indica a quantidade de registros que são classificados como itens que fazem parte de uma classe de maneira incorreta. Isto é, a resposta do classificador é que o item pertence à classe, porém, o item não pertence realmente à classe.</li>
    <li><strong>TN -</strong> Um <strong>verdadeiro negativo</strong> é um resultado em que o modelo prediz corretamente a classe negativa. Em outras palavras, indica a quantidade de registros que foram classificados como negativos de maneira correta, ou seja, a resposta do classificador foi que o item não faz parte da classe e o item realmente não faz parte da classe.</li>
    <li><strong>FN -</strong> Um <strong>falso negativo</strong> é um resultado em que o modelo prediz incorretamente a classe negativa. Portanto, indica a quantidade de registros que foram classificados como itens negativos de maneira incorreta, ou seja, a resposta do classificador é que o item não faz parte da classe, mas o item faz parte da classe.</li>
</ul>

<h4> MÉTRICAS DE CLASSIFICAÇÃO </h4>
<p style="text-align:justify;text-justify:inter-word"> Servem para identificar o quão bem o classificador está funcionando a contento, isto é, se está sendo eficiente na classificação. A resposta dessas métricas servem para ajustar os parâmetros (não se aplica ao KNN, mas sim a outros classificadores), ou reprocessar os dados que estão sendo utilizados para treinamento ou para avaliar se o modelo do classificador é o mais adequado para o tipo de classificação em questão.</p>

```python
from sklearn import metrics
print(metrics.classification_report(y_test,resultado))
```

<h4>ERRO DE CLASSIFICAÇÃO (MINIMIZADO)</h4>
<p>Mostra o quanto o classificador errou em suas predições. </p> 

```python
#############################################################################
## esse código LaTex é apenas para escrever a fórmula direitinho para vocês
from IPython.display import display, Math, Latex
print('Fórmula: ')
display(Math(r'e_{class} =\frac{FP+FN}{TP+TN+FP+FN}'))
#############################################################################
```

<h4>ACURÁCIA OU TAXA DE ACERTO</h4>
<p> A acurácia é uma métrica para avaliar modelos de classificação. Informalmente, acurácia é a fração de previsões que nosso modelo acertou. Formalmente, possui a seguinte definição:</p> 

```python
#############################################################################
## esse código LaTex é apenas para escrever a fórmula direitinho para vocês
print('Fórmula: ')
display(Math(r'Accuracy = \frac{\text{Número de Predições Corretas}}{\text{Número Total de Predições}} = \frac{TP+TN}{TP+TN+FP+FN} = 1 - e_{class}')) 
#############################################################################
```

<h4>PRECISÃO</h4>

<p style="text-align:justify;text-justify:inter-word">A "Precisão" do modelo tenta responder à seguinte pergunta: </p>
<div style="background-color: rgb(205,205,205); width: 550px; height: 55px; text-align: center">
<br>"Que proporção de identificações positivas estava realmente correta?"
</div>
 <p style="text-align:justify;text-justify:inter-word">A precisão é definida da seguinte forma:</p>


<div style="color: rgb(255,255,255); background-color: rgb(150,0,0); width: 700px; height: 55px; text-align: Justify; padding: 10px;">
 <p><strong>*** NOTA:</strong> Um modelo que não produz falsos positivos possui uma precisão de 1,0 (ou 100%)</p>
</div>

```python
#############################################################################
## esse código LaTex é apenas para escrever a fórmula direitinho para vocês
print('Fórmula: ')
display(Math(r'Precision = \frac{TP}{TP+FP}'))
#############################################################################
```

<h4>RECALL ("RECHAMADA"), RECUPERAÇÃO, REVOCAÇÃO OU SENSIBILIDADE</h4>

<p style="text-align:justify;text-justify:inter-word">É a probabilidade de um teste dar positivo na presença da classe. Número de exemplos classificados como pertencentes a uma classe, que realmente são daquela classe, dividido pela quantidade total de exemplos que pertencem a esta classe, mesmo que sejam classificados em outra.</p>

```python
#############################################################################
## esse código LaTex é apenas para escrever a fórmula direitinho para vocês
print('Fórmula: ')
display(Math(r'Recall = S = \frac{TP}{TP+FN}'))
#############################################################################
```

<h4>PRECISÃO E RECALL: UM CABO DE GUERRA </h4>
<p style="text-align:justify;text-justify:inter-word;"> Para avaliar completamente a eficácia de um modelo, você deve examinar a precisão e o recall. Infelizmente, ambos estão frequentemente em tensão. Ou seja, melhorar a precisão normalmente reduz a recuperação e vice-versa.</p> 

<h4>F1-SCORE</h4>
<p> É a média harmônica entre precisão e recall, dada por:

```python
#############################################################################
## esse código LaTex é apenas para escrever a fórmula direitinho para vocês
print('Fórmula: ')
display(Math(r'F1-Score = 2 * \frac{Precision*Recall}{Precision+Recall}'))
#############################################################################
```

<h3>OUTRAS MÉTRICAS</h3>

<h4>ESPECIFICIDADE</h4>

<p style="text-align:justify;text-justify:inter-word">É a probabilidade de um teste dar negativo na ausência da classe.</p>

```python
#############################################################################
## esse código LaTex é apenas para escrever a fórmula direitinho para vocês
print('Fórmula: ')
display(Math(r'E = \frac{TN}{TN+FP}'))
#############################################################################
```

<h4>FALSOS ALARMES</h4>

```python
#############################################################################
## esse código LaTex é apenas para escrever a fórmula direitinho para vocês
print('Fórmula: ')
display(Math(r'FA = \frac{FP}{TN+FP} = 1 - E'))
#############################################################################
```

<h4>VALOR PREDITIVO POSITIVO (VPP)</h4>
<p style="text-align:justify;text-justify:inter-word">É  a probabilidade de um item avaliado e com resultado positivo (pertencente à classe) realmente pertencer à classe.</p>

```python
#############################################################################
## esse código LaTex é apenas para escrever a fórmula direitinho para vocês
print('Fórmula: ')
display(Math(r'VPP = \frac{TP}{TP+FP}'))
#############################################################################
```

<h4>VALOR PREDITIVO NEGATIVO (VPN)</h4>
<p style="text-align:justify;text-justify:inter-word">É  a probabilidade de um item avaliado e com resultado negativo (não pertencente à classe) não pertencer realmente à classe.</p>


```python
#############################################################################
## esse código LaTex é apenas para escrever a fórmula direitinho para vocês
print('Fórmula: ')
display(Math(r'VPN = \frac{TN}{TN+FN}'))
#############################################################################
```

Um F1-Score de 90% não é perfeito, mas indica um alto nível de precisão e recall, o que significa que o modelo é capaz de identificar corretamente a maioria das instâncias positivas e negativas.</li>
    <li>A margem de erro na classificação das classes <i>Iris versicolor</i> e <i>Iris virginica</i> são iguais à 7%. Enquanto que não houve indícios de erros na classificação da classe <i>Iris setosa</i>.</li>
</ol>

<p style="text-align:justify;text-justify:inter-word;">Talvez seja possível melhorar a precisão da classificação com o uso de um outro tipo de classificador, como uma rede neural, por exemplo, retreinando-a até pegar os melhores resultados dos parâmetros. Entretanto, o intuito aqui é ter justamente ter essa percepção de julgamento sobre os dados, sobre o classificador e sobre os resultados. Compreendendo, é claro, o uso das métricas e, também, das demais técnicas apresentadas anteriormente entendendo os dados que estão sendo trabalhados.</p>
